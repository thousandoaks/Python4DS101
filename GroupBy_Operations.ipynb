## GroupBy Operations

GroupBy Operations follow the so-called Split-Apply-Combine paradigm whereby a `DataFrame` is: (1) split into groups, (2) a function is applied to each group and (3) the result are combined into a new `DataFrame` object.



<img src='https://pandas.pydata.org/docs/_images/06_groupby.svg'>

For instance in the following example we split the titanic dataset into groups based on gender, we compute the mean of each group.

Based on the results we notice that:

* Females on average are younger than their male counterparts
* Females on average paid higher fares than their male counterparts.

```
titanic_Age_Fare_Gender.groupby('Sex').mean()
```

We could have applied other functions to each group, for instance `sum` or `min` or `max`

```
titanic_Age_Fare_Gender.groupby('Sex').sum()
```

```
titanic_Age_Fare_Gender.groupby('Sex').min()
```

```
titanic_Age_Fare_Gender.groupby('Sex').max()
```
